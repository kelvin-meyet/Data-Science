# DIAGNOSTIC PREDICTION OF DIABETES AMONG PIMA INDIAN HERITAGE

```python
import pandas
from sklearn import model_selection
from sklearn.ensemble import AdaBoostClassifier
```

```python
dataframe = pandas.read_csv('pima-indians-diabetes.csv')
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
seed = 7
num_trees = 30
```

# Applying Boosting to Classification Models

```python
#Build Classifiers with Adaboost
#Adaboost classifiers use decision trees classifiers as default
#Passed the model within a cross validation function to evaluate results with a cv technique(train data splitted into 10 consecutive folds),  
#Each fold will be used as evaluation set while remaining 9 fold used as training set
kfold = model_selection.KFold(n_splits=10,shuffle=True, random_state=seed)
model = AdaBoostClassifier(n_estimators=num_trees)
results = model_selection.cross_val_score(model,X,Y,cv=kfold)
print(results)
print(results.mean())
#accuracy of 76% with adaboost
```

# Random Forest with C-V splitting technique 

```python
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()
print(iris_data)
data_input = iris_data.data #extract input data
data_output = iris_data.target #extract target data
```

```python
print(data_input)
print(data_output)
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True)
```

```python
print("Train Set           Test Set      ")
for train_set, test_set in kf.split(data_input):
    print(train_set, test_set)
```

```python
#intialize Random forest Classifier

from sklearn.ensemble import RandomForestClassifier
rf_class = RandomForestClassifier(n_estimators=10)
```

```python
#import cross validation score library from sklearn
from sklearn.model_selection import cross_val_score
print(cross_val_score(rf_class, data_input, data_output, scoring='accuracy',cv=10))
```

```python
accuracy = cross_val_score(rf_class, data_input, data_output, scoring='accuracy', cv=10).mean()*100
print('Accuracy of Random Forest is:',accuracy)
```
